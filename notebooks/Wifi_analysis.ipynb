```python
# 2. Function: Save Data to CSV
def save_signal_data(data, index_values, level_tag="L"):
    for wing in data['wifi']:
        df_wing = pd.DataFrame({
            'wifi': data['wifi'][wing],
            'mobile_net': data['mobile_net'][wing]
        }, index=index_values[wing])

        filename = f"{level_tag}_{wing}_signal_strength.csv"
        df_wing.to_csv(filename, index=True)
        print(f"‚úÖ Saved: {filename}")
```

```python
# 5. Load DataFrames from CSV
def load_wing_data(level_tag="L4", wings=['a','b','c','d','e','f','g','h']):
    df_dict = {}
    for wing in wings:
        filename = f"/content/Dataset/{level_tag}/{level_tag}_{wing}_signal_strength.csv"
        if os.path.exists(filename):
            df_dict[wing] = pd.read_csv(filename, index_col=0)
            print(f"üìÇ Loaded: {filename}")
        else:
            print(f"‚ùå File not found: {filename}")
    return df_dict
```

```python
# 6. Visualization Function
def plot_signal_comparison(df_dict, wing):
    if wing not in df_dict:
        print(f"‚ö†Ô∏è Wing '{wing}' not found.")
        return
    data = df_dict[wing]
    data.plot(kind='bar', color=['#1f77b4', '#ff7f0e'])
    plt.title(f"Signal Strength Comparison - Wing {wing.upper()}")
    plt.ylabel("Signal Strength (dBm)")
    plt.xlabel("Measurement")
    plt.xticks(rotation=45)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
import plotly.graph_objects as go
import plotly.subplots as sp

# Load Level 4 data
df = load_wing_data("L4")

# Define wings list
wings = ['e', 'f', 'g', 'h', 'a', 'b', 'c', 'd']

# Create a Plotly subplot figure (rows = number of wings)
fig = sp.make_subplots(
    rows=len(wings), cols=1, shared_xaxes=True,
    subplot_titles=[f"Wing {wing.upper()}" for wing in wings]
)

# Loop through each wing to create a bar chart
for i, wing in enumerate(wings):
    wifi_signal = df[wing]['wifi']
    mobile_signal = df[wing]['mobile_net']

    # Create bar traces for Wi-Fi and Mobile Network signal strength
    trace_wifi = go.Bar(
        x=df[wing].index, y=wifi_signal, name="Wi-Fi",
        marker=dict(color='#BE3144'),
        hoverinfo='x+y+name',
    )

    trace_mobile = go.Bar(
        x=df[wing].index, y=mobile_signal, name="Mobile Network",
        marker=dict(color='#E17564'),
        hoverinfo='x+y+name',
    )

    # Add traces to the subplot
    fig.add_trace(trace_wifi, row=i+1, col=1)
    fig.add_trace(trace_mobile, row=i+1, col=1)

# Update layout for the figure
fig.update_layout(
    title="Signal Strength Comparison: Wi-Fi vs Mobile Network",
    height=600 * len(wings),
    xaxis_title="Measurement Point",
    yaxis_title="Signal Strength (dBm)",
    barmode='group',
    showlegend=True,
    legend=dict(x=0.8, y=1),
    # plot_bgcolor='#222222',  # Dark background for better contrast
    # paper_bgcolor='#222222',  # Dark background for paper as well
    # font=dict(color='white'),
    # hovermode="x unified",  # Hover mode for all traces
)

# Show the interactive plot
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Level 3 data
df = load_wing_data("L3")


# Calculate the average signal strength for Wi-Fi and Mobile Network for each wing
avg_wifi = {wing: df[wing]['wifi'].mean() for wing in df}
avg_mobile_net = {wing: df[wing]['mobile_net'].mean() for wing in df}

# Convert the averages to a DataFrame for easier plotting
avg_df = pd.DataFrame({
    'Wi-Fi': avg_wifi,
    'Mobile Network': avg_mobile_net
}).T

# Plot the average signal strengths
plt.figure(figsize=(12, 7))

# Use a distinct color for Wi-Fi and Mobile Network without using a palette
avg_df.plot(kind='bar', color=['#BE3144', '#E17564'], edgecolor='white', width=0.8, linewidth=1.5)

# Title and labels
plt.title('Average Wi-Fi vs Mobile Network Signal Strength Across Wings', fontsize=20, color='white', pad=20)
plt.ylabel('Average Signal Strength (dBm)', fontsize=16, color='white', labelpad=10)
plt.xlabel('Wing', fontsize=16, color='white', labelpad=10)
plt.xticks(rotation=45, fontsize=14, color='white', ha='right')
plt.yticks(fontsize=14, color='white')

# Adjust gridlines for readability and set background color
plt.grid(True, linestyle='--', color='white', alpha=0.1)
plt.gca().set_facecolor('#2E2E2E')  # Darker background color for the plot
plt.gcf().set_facecolor('#222222')  # Dark background for the whole figure

# Adjust legend with improved appearance
plt.legend(
    loc='upper left',
    fontsize=14,
    title='Signal Type',
    title_fontsize=16,
    facecolor='#3E3E3E',
    edgecolor='white',
    framealpha=0.7,
    labelspacing=1.2,
    labels=['Wi-Fi', 'Mobile Network']
)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Load Level 3 data
df = load_wing_data("L3")
# Combine data for violin plot
violin_data = {
    'wifi': [],
    'mobile_net': [],
    'wing': []
}

# Flatten the data for each wing
for wing in df:
    wifi_values = df[wing]['wifi']
    mobile_net_values = df[wing]['mobile_net']

    # Append data for this wing
    violin_data['wifi'].extend(wifi_values)
    violin_data['mobile_net'].extend(mobile_net_values)
    violin_data['wing'].extend([wing] * len(wifi_values))  # Repeat the wing label for each measurement

# Create a DataFrame for plotting
violin_df = pd.DataFrame(violin_data)

# Create the violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='wing', y='wifi', data=violin_df, color='#BE3144', inner="quart", linewidth=1.2)
sns.violinplot(x='wing', y='mobile_net', data=violin_df, color='#E17564', inner="quart", linewidth=1.2)

# Apply custom title and labels with improved contrast
plt.title('Distribution of Wi-Fi and Mobile Network Signal Strength Across Wings', fontsize=16, color='white')
plt.xlabel('Wing', fontsize=14, color='white')
plt.ylabel('Signal Strength', fontsize=14, color='white')

# Improve layout and visibility
plt.tight_layout()

# Set a better style for the legend
plt.legend(title="Signal Type", fontsize=12, loc='best', facecolor='white', edgecolor='black', framealpha=0.7)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate

# Generate a grid for the contour plot
# Assuming you're using the indices for X and Y, and Wi-Fi values for Z
# Load Level 3 data
df = load_wing_data("L3")
# Create a list to store the coordinates and Wi-Fi values
coordinates = []
wifi_values = []

# Collect the data for all wings
for wing in df:
    x = np.arange(len(df[wing]))  # Measurement points (x-axis)
    y = np.random.normal(0, 0.1, len(x))  # Add small random noise to break flatness
    wifi_data = df[wing]['wifi']

    # Append the data
    coordinates.extend(zip(x, y))
    wifi_values.extend(wifi_data)

# Convert to numpy arrays
coordinates = np.array(coordinates)
wifi_values = np.array(wifi_values)

# Interpolate data into a 2D grid
grid_x, grid_y = np.meshgrid(np.linspace(coordinates[:, 0].min(), coordinates[:, 0].max(), 100),
                             np.linspace(coordinates[:, 1].min(), coordinates[:, 1].max(), 100))

# Perform griddata interpolation to get Z (Wi-Fi signal strength) values on the grid
grid_z = scipy.interpolate.griddata(coordinates, wifi_values, (grid_x, grid_y), method='cubic')

# Create the plot
plt.figure(figsize=(12, 6))

# Scatter plot of the data points (Wi-Fi signal strength)
for wing in df:
    x = np.arange(len(df[wing]))  # Measurement indices as x
    y = np.random.normal(0, 0.1, len(x))  # Add small random noise to break flatness
    plt.scatter(x, y, c=df[wing]['wifi'], cmap='coolwarm', edgecolors='black', label=f'Wi-Fi - Wing {wing.upper()}')

# Plot contours
plt.contour(grid_x, grid_y, grid_z, levels=15, cmap='coolwarm', alpha=0.5)

plt.title(f'Contour Plot of Wi-Fi Signal Strength')
plt.xlabel('wings')
plt.ylabel('locations')
plt.colorbar(label='Wi-Fi Signal Strength')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
